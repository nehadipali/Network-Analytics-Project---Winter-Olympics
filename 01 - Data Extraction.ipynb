```python
import pandas as pd
import tweepy
```

```python
api_key = #enter api key
secret_key = #enter secret key
access_token = #enter access token
access_token_secret = #enter secret access token

auth = tweepy.OAuthHandler(api_key,secret_key)
auth.set_access_token(access_token,access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

Scraped 20,000 tweets instead of 5,000 since we had around 3,500 unique users when we scraped 5,000 tweets.

```python

eileenGuTweets = pd.DataFrame(columns = ['Content Type','Tweet','A_username','A_followers','A_following','A_listed','A_posts','A_location','B_text','B_username',
                                            'B_followers','B_following','B_listed','B_posts','B_location'])
tweet_info = []
hashtag = 'Eileen Gu'
numTweets = 30000

for tweet in tweepy.Cursor(api.search_tweets,
                           q=hashtag,
                           count=100,
                           result_type="recent",
                           include_entities=True).items(numTweets):
    try:
        tweet_info.append(tweet)
        a_username = tweet._json['user']['screen_name']
        if tweet.text[:2] == 'RT': # they are retweeting
            b_handle = tweet._json['retweeted_status']['user']['screen_name']
            b_followers = tweet._json['retweeted_status']['user']['followers_count']
            b_following = tweet._json['retweeted_status']['user']['friends_count']
            b_listed = tweet._json['retweeted_status']['user']['listed_count']
            b_posts = tweet._json['retweeted_status']['user']['statuses_count']
            b_location = tweet._json['retweeted_status']['user']['location']
            content = 'Interaction'
        elif False:
            pass
        else:
            b_username = a_username
            content = 'Self'
        tweetActual = tweet.text
        a_followers = tweet._json['user']['followers_count']
        a_following = tweet._json['user']['friends_count']
        a_listed = tweet._json['user']['listed_count']
        a_posts = tweet._json['user']['statuses_count']
        a_location = tweet._json['user']['location']
        

        eileenGuTweets = eileenGuTweets.append(pd.Series({'Content Type':content,'Tweet':tweetActual,'A_username':a_username,'A_followers':a_followers,
                                          'A_following':a_following,'A_listed':a_listed,'A_posts':a_posts,'A_location':a_location,
                                          'B_username':b_username,'B_followers':b_followers,
                                          'B_following':b_following,'B_listed':b_listed,
                                          'B_posts':b_posts,'B_location':b_location}), ignore_index = True)
    except:
        continue

eileenGuTweets.head()
```

```python
eileenGuTweets.to_excel("eileenGuTweets.xlsx")  
```
